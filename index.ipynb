### Your Task

Your task is to create a bar graph of the 25 most common words in the Shakespeare's Macbeth.  


A common python programming pattern to counting objects, produce histograms, or update statistics is to make calls to a dictionary as you iterate through a list. For example, given a list of words, you can create a dictionary to store counts and then iterate through the list of words, checking how many times each word has appeared using your dictionary, and updating the dictionary count now that you've seen that word again. The `dictionary.get()` method is very useful in doing this.
```python
# Your code here

# Pseudo-code Outline
# Split the transcript into words
# Create a dictionary
# Iterate through the text of Macbeth
# Update word counts
# Create Bar Graph
# Include descriptive titles and 
import copy  # to make a deep copy of lists



def remove_punct(wpunct): #turn punctuation characters into spaces
    wopunct = copy.deepcopy(wpunct)
    punct = ["*", ".", ",", "?", ":", ";", "!", "\n", "\r", "(", ")"]
    for i in range(len(wpunct)): 
        for p in punct:
            if wopunct[i] == p:
                wopunct[i] = " " #turn every punctuation mark into a space
    return wopunct
    
def remove_spaces(wspaces): # collapse double spaces into single spaces
    wospaces = copy.deepcopy(wspaces)
    leng = len(wospaces)
    for i in range(leng): 
        if i < (leng - 1):
            if wospaces[i] == " " and wospaces[i + 1] == " ": #if a double space
                del wospaces[i + 1] # delete the second consecutive space
                leng -= 1 # and shorten the length of the index
    return wospaces
```

```python
macbeth_list = list(macbeth)        
macbeth_list_wopunct = remove_punct(macbeth_list)        
len(macbeth_list_wopunct)
macbeth_list_wopunct[:500]
```

```python
macbeth_list_2 = remove_spaces(macbeth_list_wopunct) # Keep calling remove_spaces until
macbeth_list_3 = remove_spaces(macbeth_list_2) #all double spaces are gone.
macbeth_list_4 = remove_spaces(macbeth_list_3) # I don't know yet how to iterate
macbeth_list_5 = remove_spaces(macbeth_list_4) # this part.
macbeth_list_6 = remove_spaces(macbeth_list_5)
len(macbeth_list_6)
```

```python
string_wopunct_wospace = "".join(macbeth_list_6) # Put clean list back into a string

lower_string = string_wopunct_wospace.lower() # make all letters lower case

clean_script = lower_string.split(" ") # Separate clean string back into a list of words divided by spaces.

unique_words_set = set(clean_script) #  Then get unique words in a set.

unique_words = list(unique_words_set) #list of unique words in macbeth.

counts = dict.fromkeys(unique_words, 0)  # a dictionary to count instances of each word

counts
```

```python
for word in unique_words: # loop over each word in the unique words list
    for item in clean_script: # loop over each word in the clean script
        if word == item:
            counts[word] += 1

counts_inorder = sorted(counts, key=lambda x: counts[x]) #order unique words by frequency
top_25_words = counts_inorder[-25:-1] #list of only 25 highest-frequency words
top_25_counts = list(map(lambda count: counts[count], top_25_words)) #list of 
print(top_25_words)
print(top_25_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(top_25_words, top_25_counts)
plt.xlabel("Top 25 Words")
plt.ylabel("Frequency Word Appears in Text")
plt.title("Top 25 Words Appearing in Macbeth")
plt.legend()
plt.show()
```
